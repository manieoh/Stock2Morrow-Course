```python
import pandas as pd

#สร้าง List ของ Time โดยกำหนดให้เวลาเริ่มจาก 01/01/2020 จำนวน 5 ครั้ง และเพิ่มทีล่ะ 12 ชั่วโมง 
ind = pd.date_range('01/01/2020', periods = 5, freq ='12H') 
   
#สร้าง DataFrame ที่มี Columns A,B,C,D และกำหนด Index เป็น ind
df = pd.DataFrame({"A":[1, 2, 3, 4, 5],  
                   "B":[10, 20, 30, 40, 50], 
                   "C":[11, 22, 33, 44, 55], 
                   "D":[12, 24, 51, 36, 2]},  
                    index = ind) 
# แสดง DataFrame
df 
```

```python
df.shift(2, axis=0) #เลื่อน Rows ลงไปจำนวน 2 ช่อง
```

```python
df.shift(-2, axis=0) #เลื่อน Rows ขึ้นไปจำนวน 2 ช่อง
```

```python
df.shift(2, axis=1) #เลื่อน Columns  ไปทางขวาจำนวน 2 ช่อง
```

```python
df.shift(-2, axis=1) #เลื่อน Columns  ไปทางซ้ายจำนวน 2 ช่อง
```
