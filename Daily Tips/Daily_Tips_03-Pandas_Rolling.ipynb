```python
def count_lower(x):
    np_list = np.array(x)
    condition = np_list<x[-1]
    count = condition.sum()
    return count

def count_higher(x):
    np_list = np.array(x)
    condition = np_list>x[-1]
    count = condition.sum()
    return count

df['number_of_lower'] = df.close.rolling(5).apply(lambda x: count_lower(x), raw=True)
df['number_of_higher'] = df.close.rolling(5).apply(lambda x: count_higher(x), raw=True)
df.head(15)
```
