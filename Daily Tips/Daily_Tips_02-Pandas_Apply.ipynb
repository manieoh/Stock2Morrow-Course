```python
import pandas as pd
import numpy as np

#สร้าง List ของ Time โดยกำหนดให้เวลาเริ่มจาก 01/01/2020 จำนวน 5 ครั้ง และเพิ่มทีล่ะ 12 ชั่วโมง 
ind = pd.date_range('01/01/2020', periods = 5, freq ='12H') 
   
#สร้าง DataFrame ที่มี Columns A,B,C,D และกำหนด Index เป็น ind
df = pd.DataFrame({"A":[1, 2, 3, 4, 5],  
                   "B":[10, 20, 30, 40, 50], 
                   "C":[11, 22, 33, 44, 55], 
                   "D":[12, 24, 51, 36, 2]},  
                    index = ind) 
# แสดง DataFrame
df 
```

```python
df.apply(np.sqrt) #การสั่ง Apply ให้แต่ล่ะช่องคำนวณ Square Root
```

```python
df.apply(np.sum, axis=0) #การสั่ง Apply โดยใช้ axis=0 เพื่อให้ sum ใน Columns
```

```python
df.apply(np.sum, axis=1) #การสั่ง Apply โดยใช้ axis=1 เพื่อให้ sum ใน Rows
```

```python
df['SUM'] = df.apply(np.sum, axis=1) #หรือจะใช้ Apply ในการช่วยสร้างข้อมูลใหม่ก็ได้
df
```

```python
#เพิ่มเติม lambda เพื่อให้สร้างฟังก์ชั่นการคำนวณได้ด้วยตัวเอง
df['SUM2'] = df['SUM'].apply(lambda x: True if x>70 else False)
df
```

```python
#เพิ่มเติม lambda เพื่อให้สร้างฟังก์ชั่นการคำนวณได้ด้วยตัวเอง
df['SUM3'] = df['SUM2'].apply(lambda x: 1 if x else -1)
df
```
