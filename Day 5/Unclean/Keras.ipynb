# Keras

```python
from keras.layers import Dense, Input
from keras.models import Model
```

```python
data = df.copy()
```

```python
data['label'] = df['close'].rolling(5, center=True).mean().diff().shift(-1)
data.drop(['momentum_kama', 'trend_psar_up', 'trend_psar_down'], axis=1, inplace=True)
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
data = pd.DataFrame(mms.fit_transform(data), columns=data.columns, index=data.index)
```

```python
data
```

```python
X = data.drop(['label'], axis=1)
y = data['label']
```

```python
X_train = X.iloc[:-5000][select]
y_train = y.iloc[:-5000]

X_test = X.iloc[-5000:][select]
y_test = y.iloc[-5000:]

close = X.iloc[-5000:]['close']
```

```python
X_train
```

```python
input_layer = Input(shape=(37, ))
```

```python
dense_layer1 = Dense(128, activation='relu')(input_layer)
```

```python
dense_layer2 = Dense(64, activation='relu')(dense_layer1)
```

```python
dense_layer3 = Dense(32, activation='relu')(dense_layer2)
```

```python
output = Dense(1, activation='sigmoid')(dense_layer3)
```

```python
model = Model(inputs = input_layer, outputs = output)
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='mean_absolute_error')
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=1024)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred.reshape(-1)
```

```python
mean_absolute_error(y_test, y_pred.reshape(-1))
```

```python
y_train.median()
```

## Convolution Net

Convolution and Recurrent Layers take 3D Tensors.

```python
data.drop(['label'], axis=1).iloc[0:24, :].values
```

```python
from tqdm import tqdm
```

```python
arr = []
for i in tqdm(range(0, 1000)):
    arr.append(data.drop(['label'], axis=1).iloc[i:i+10, :].values)
```

```python
X = np.array(arr)
```

```python
X
```

```python
y = data['label'].iloc[9:1009].values
```

```python
from keras.layers import Conv1D, Flatten
```

```python
X.shape
```

```python
input_layer = Input(shape=(10,80, ))
```

```python
conv1 = Conv1D(10, 4, activation='relu')(input_layer)
```

```python
conv2 = Conv1D(10, 4, activation='relu')(conv1)
```

```python
flat = Flatten()(conv2)
```

```python
fc = Dense(32, activation='relu')(flat)
```

```python
output = Dense(1, activation='sigmoid')(fc)
```

```python
model = Model(input_layer, output)
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='mean_absolute_error')
```

```python
model.fit(X, y, epochs=50, batch_size=32)
```
