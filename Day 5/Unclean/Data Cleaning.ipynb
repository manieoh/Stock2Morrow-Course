# Data Imputation Strategies

1. Drop NA
2. Impute with Simple Value
3. Impute with Statistics (Mean, Median, ...)
4. Impute with Nearest Neighbor

```python
df.dropna()
```

```python
df.dropna().info()
```

```python
df.fillna(0)
```

```python
df.mean()
```

```python
df.fillna(df.mean())
```

```python
df.fillna(df.median())
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```
